### Dynamic Amplification Factor
\begin{equation}
DMF = \frac{1}{\sqrt{(1-\beta^2)^2 + (2\zeta\beta)^2}}
\end{equation}

The dynamic amplification Factor tell us about the effect of the forcing frequency on the response of the strcuture. It depends on 
1. Forcing Frequency
2. Damping
on the response of the structure

Assuming a frequency ratio between 0 and 2.
Assuming 5% damping in the system

```python
import numpy as np
import math
import matplotlib.pyplot as plt
import seaborn as sns

max_B = 3
delta_B = 0.01
nPoints = max_B/delta_B
B = np.linspace(0,max_B,int(nPoints))

xi = 0.05
```

```python
dmf = 1/(np.sqrt((1-B**2)**2 + (2*xi*B)**2))
```

```python
sns.set_style('whitegrid')
plt.figure(figsize=(12,4))
plt.plot(B,dmf)
plt.title("Forcing Frequency v/s Dynamic Amplification Factor")
plt.xlabel('Frequency Ratio')
plt.ylabel("Dynamic Amplification Factor")
plt.ylim(0,max(dmf)+2)
plt.xlim(0,max_B)
plt.vlines(1.0,0,max(dmf),colors='red',lw = 0.75)
```

#### Important Points - 

##### 1.It is observed that the DMF increases abruptly as the forcing frequency gets near to the natural frequency of the system.
##### 2. The DMF is zero at zero forcing frequency meaning that if forcing frequecny is 0 indicate the deflection will be same as the static response of the structure.
##### 3. The DMF nears 0 as the time increases meaning that as the forcing frequecny becomes too high then the displacement of the system is essentially zero.

## Effects of various damping ratios on the DMF

```python
Xi = [0.05,0.1,0.2,0.50,0.75,1.0]

# plotting
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,2,1.5])
axes.set_xlabel('Frequency Ratio $\\beta$',{
    'fontsize': 16,
    'fontweight': 'bold',
    'color': 'black'
})
axes.set_ylabel('Dynamic Magnification Factor',{
    'fontsize': 16,
    'fontweight': 'bold',
    'color': 'black'
})
axes.set_title('Dynamic Magnification Factor',{
    'fontsize': 24,
    'fontweight': 'bold',
    'color': 'black'
})
axes.set_xlim(0,2)
axes.set_ylim(0,12)
axes.grid()

for xi in Xi:
    dmf = 1/(np.sqrt((1-B**2)**2 + (2*xi*B)**2))
    axes.plot(B,dmf,label = f'Damping Ratio {xi*100} %',lw=2)
axes.plot([1.0,1.0],[-10,20])
plt.legend(fontsize =15)   
```
